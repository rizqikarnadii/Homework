## 5. Show info table of the DF like this

```python
cc_desc = []

for i in creditcard.columns:
    cc_desc.append([i, creditcard[i].dtypes,
                         creditcard[i].isna().sum(),
                         round((((creditcard[i].isna().sum())/(len(creditcard)))*100),2),
                         creditcard[i].nunique(),
                         creditcard[i].drop_duplicates().sample(2).values])
    
pd.DataFrame(cc_desc, columns = ['dataFeatures', 'dataType', 'null', 'nullPct', 'unique', 'uniqueSample'])
```
